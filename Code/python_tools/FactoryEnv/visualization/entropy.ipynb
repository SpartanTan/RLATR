```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.special import entr

# Range of standard deviations to consider
sigmas = np.linspace(0.1, 3, 100)

# Calculate entropy for each sigma
# entropies = [norm.entropy(scale=sigma) for sigma in sigmas]
entropies = 0.5 + 0.5 * np.log(2 * np.pi * np.e * np.square(sigmas))

# Plotting
plt.plot(sigmas, entropies)
plt.xlabel('Standard Deviation (Ïƒ)')
plt.ylabel('Entropy')
plt.title('Entropy of Normal Distribution vs Standard Deviation')
plt.grid(True)
plt.show()
```
