```python
import numpy as np
import matplotlib.pyplot as plt

# Set the mean (lambda) value for the Poisson distribution
lam = 0.25  # You can change this value as needed

# Generate Poisson distribution values using NumPy
data = np.random.poisson(lam, 1000)

# Plotting the distribution
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Add a plot title and labels
plt.title('Poisson Distribution ($\mu$ = {})'.format(lam))
plt.xlabel('Number of Events')
plt.ylabel('Probability')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Computational times for the three algorithms
times = [1.4170680046081543, 3.5816025733947754, 6.444879531860352, 11.155172109603882]

# Algorithm names (for labeling purposes, you can replace these with actual algorithm names)
algorithms = ['Method-in-term-of\n-obstacles-and-walls','Emulator with \nresoluion 12\u00B0', 'Emulator with \nresoluion 6\u00B0', 'Emulator with \nresoluion 3\u00B0']

# Creating a bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(algorithms, times, color=['blue', 'green', 'red'])
# Adding the computational times above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

# Adding titles and labels
plt.title('Computational Time Comparison of Different Laser Emulator')
plt.xlabel('Methods')
plt.ylabel('Computational Time (seconds)')
plt.grid()
# Showing the plot
plt.show()
```
